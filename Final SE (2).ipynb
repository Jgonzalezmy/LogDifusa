```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

## Universo de discurso

```python
#Nivel de Contaminacion
x_nivel = np.arange(1,4,1)

#Tipo de Contaminacion
x_tipo = np.arange(1,4,1)
```

```python
x_nivel
```

```python
x_tipo
```

## Funciones

```python
nivel_V = fuzz.trimf(x_nivel,[1,1,1])
nivel_N = fuzz.trimf(x_nivel,[2,2,2])
nivel_R = fuzz.trimf(x_nivel,[3,3,3])

tipo_V = fuzz.trimf(x_tipo,[1,1,1])
tipo_N = fuzz.trimf(x_tipo,[2,2,2])
tipo_R = fuzz.trimf(x_tipo,[3,3,3])
```

## Visualizacion

```python
fig, (ax0,ax1) = plt.subplots(nrows=2, figsize=(5,10))

ax0.plot(x_nivel,nivel_V,'g', linewidth=2, label='ALERTA VERDE')
ax0.plot(x_nivel,nivel_N,'y', linewidth=2, label='ALERTA NARANJA')
ax0.plot(x_nivel,nivel_R,'r', linewidth=2, label='ALERTA ROJA')
ax0.set_title('NIVELES DE CONTAMINACION EN MEDELLIN')
ax0.legend()

ax1.plot(x_tipo,tipo_V,'g', linewidth=2, label='BAJO')
ax1.plot(x_tipo,tipo_N,'y', linewidth=2, label='MEDIO')
ax1.plot(x_tipo,tipo_R,'r', linewidth=2, label='ALTO')
ax1.set_title('TIPOS DE CONTAMINACION')
ax1.legend()
```

## Fuzificacion

```python
n = int(input('EL NIVEL DE CONTAMINACION DETECTADO ES ' ))
x_nivel_V = fuzz.interp_membership(x_nivel,nivel_V,n)
x_nivel_N = fuzz.interp_membership(x_nivel,nivel_N,n)
x_nivel_R = fuzz.interp_membership(x_nivel,nivel_R,n)


t = int(input('EL TIPO DE CONTAMINACION ' ))
x_tipo_V = fuzz.interp_membership(x_tipo,tipo_V,t)
x_tipo_N = fuzz.interp_membership(x_tipo,tipo_N,t)
x_tipo_R = fuzz.interp_membership(x_tipo,tipo_R,t)
```

## Reglas

```python
#Visualizar
fig, ax0 = plt.subplots(figsize=(8,6))
```

```python
regla1 = np.fmax(x_nivel_V,x_tipo_V)
regla1
```

```python
regla2 = np.fmin(x_nivel_N,x_tipo_N)
regla2
```

```python
regla3 = np.fmin(x_nivel_R,x_tipo_R)
regla3
```

```python
result = np.fmax(regla1, np.fmax(regla2, regla3))
result
```
